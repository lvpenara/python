```python
import pandas as pd
from datetime import date
from datetime import datetime 
from datetime import timedelta
```

```python
fecha_ini = datetime(2020,3,1)
fecha_fin = datetime(2020,9,30)
days = (fecha_fin - fecha_ini).days
```

```python
for i in range(0,days+1):
    fecha_aux = fecha_ini + timedelta(days=i)
    print('dia:',fecha_aux.day,' ,mes:',fecha_aux.month, ' ,año:',fecha_aux.year )
```
