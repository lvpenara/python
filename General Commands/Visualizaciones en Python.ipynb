```python
#Generando datos de una densidad normal estándar
data1=np.random.normal(loc=1.0, scale=0.5,size=1000)
data2=np.random.normal(loc=1.5, scale=0.1,size=500)
data3=np.random.normal(loc=0.5, scale=0.7,size=500)

print(data1.shape)
print(type(data1))
data1[5:10]
#Cambiar de forma se hace con un reshape


```

```python
#Creando un histograma
plt.hist(data1,bins=14);
plt.show()
```

```python
#Creando un histograma
#alpha es la transparencia
plt.hist(data2,bins=14,alpha=0.4,label='data2');
plt.hist(data3,bins=14,alpha=0.4,label='data3');
plt.legend()
plt.show()
#Haciendo dos histogramas, uno al lado del otro para poder comparar
```

```python
#Generación de datos
x = np.linspace(0,10,10)
y = x**2
plt.plot(x,y,'x-', label='Cuadrática')
plt.plot(x,x,'o-', label='Lineal')

plt.title('Relación Cuadrada y Lineal')
#plt.legend()   #para ubicar en el mejor sitio del gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```
