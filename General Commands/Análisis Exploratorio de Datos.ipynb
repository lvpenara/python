```python
#ejemplo de cÃ¡lculo de percentiles con numpy
np.percentile([1,2,3,4,5,6],[25,50,75])
```

```python
#Definiendo la funciÃ³n para poder encontrar atÃ­picos
def test_tukey(df,columna,extremo=False):
    q1, q3 = np.percentile(df[columna],[25,75])
    iqr=q3-q1
    c = 3 if extremo else 1.5
    filtro = (df[columna]<= (q3+iqr*c)) | (df[columna]>=(q1-iqr*c))
    return df[filtro]
```

```python
df_sp_out=test_tukey(df,'SalePrice')
df_out=test_tukey(df_sp_out,'GrLivArea');
df_out.plot(kind='scatter',x='GrLivArea',y='SalePrice');
df.plot(kind='scatter',x='GrLivArea',y='SalePrice')
```
